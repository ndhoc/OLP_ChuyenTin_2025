# C. Hình vuông

time limit per test: 2 seconds

memory limit per test: 512 megabytes

Trường Đại học Sư phạm Kỹ thuật TPHCM nổi tiếng với rất nhiều cây xanh trong khuôn viên trường, khuôn viên trường có thể xem như một mặt phẳng hai chiều vô hạn. Các cây trong trường được đánh số từ $1$ đến $N$ để dễ quản lý.

Bảo được giao nhiệm vụ làm người quản lý và chăm sóc các cây xanh này. Bảo vừa được giao nhiệm vụ là xây dựng hàng rào bao quanh các cây xanh trong đoạn $[L,R]$ nhất định. Trước khi bắt đầu xây dựng, Bảo cần tính toán trước chi phí cần thiết. Chi phí cần thiết để xây dựng hàng rào bao quanh các cây xanh trong đoạn $[L,R]$ là cạnh của hình vuông nhỏ nhất, sao cho các cạnh của nó song song với trục tọa độ và các cây xanh trong đoạn $[L,R]$ nằm bên trong hoặc trên cạnh của hình vuông này.

Với một đoạn $[L,R]$, Bảo có thể bỏ qua tối đa $1$ cây xanh (bỏ qua một cây xanh trong việc tính chi phí chắc không ai để ý đâu nhỉ?).

Bảo có tổng cộng $Q$ đoạn cần tính toán chi phí, nhiệm vụ của bạn là giúp Bảo tính toán những chi phí này.

### **Input**

*   Dòng đầu tiên gồm hai số nguyên dương $N$ và $Q$.

*   N dòng tiếp theo, mỗi dòng gồm hai số nguyên $x_i,y_i$ biểu diễn tọa độ của cây thứ $i$.

*   $Q$ dòng tiếp theo, mỗi dòng gồm hai số nguyên $L,R$.

### **Output**

*   Với mỗi truy vấn $[L,R]$, in ra một số nguyên là độ dài cạnh hình vuông nhỏ nhất cần xây dựng.

#### **Constraints**

*   $2≤N≤105$

*   $1≤Q≤105$

*   $−10^9≤x_i,y_i≤10^9$

*   $1≤L<R≤N$

### **Scoring**

*   Subtask 1 (20% số điểm): $N,Q≤100$

*   Subtask 2 (40% số điểm): $N,Q≤5000$

*   Subtask 3 (40% số điểm): Không còn ràng buộc nào khác.

### **Examples**

**input**

3 2

1 0

0 1

1000 1

1 3

2 3

**output**

1

0

**input**

4 2

0 0

1000 1000

300 300

1 1

1 3

2 4


**output**

300

299

### **Note**

**Giải thích ví dụ 1:**

*   Với truy vấn $1$, Bảo nên bỏ qua cây thứ $3$ trong việc tính chi phí.

*   Với truy vấn $2$, do chỉ có $2$ cây nên việc bỏ qua cây nào cũng dẫn tới chi phí như nhau.
