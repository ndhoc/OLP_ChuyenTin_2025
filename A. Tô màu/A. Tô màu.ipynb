# A. Tô màu

time limit per test: 1 second

memory limit per test: 256 megabytes


Bảo vừa được tặng một đồ thị gồm $N$ đỉnh được đánh số từ $1$ tới $N$. Ban đầu $N$ đỉnh này chưa tô màu gì hết, Bảo muốn tô cho $N$ đỉnh này hai màu đen hoặc trắng.

Vì Bảo rất khắt khe với việc tô màu các đỉnh của đồ thị này, nên Bảo muốn tô màu sao cho đảm bảo được toàn bộ $M$ điều kiện Bảo đề ra.

Một điều kiện có dạng $u$ $v$ $c$, với $c$ là số đỉnh có màu đen trong số hai đỉnh $u$ và $v$. Bạn hãy giúp Bảo tô màu đồ thị này hoặc thông báo với Bảo điều này là không thể.

### **Input**

*   Dòng đầu tiên gồm hai số nguyên dương $N,M$
*   $M$ dòng tiếp theo, mỗi dòng gồm 3 số nguyên dương $u$ $v$ $c$



#### **Constraints**

*   $1≤N,M≤100000$
*   $1≤u,v≤N,u≠v$
*   $c∈{0,1,2}$



### **Output**

*   Nếu có thể tô màu sao cho đảm bảo $M$ điều kiện, in ra $N$ số nguyên, với số nguyên thứ $i$ là $0$ nếu đỉnh $i$ tô màu trắng, ngược lại, số nguyên thứ $i$ là 1 nếu đỉnh $i$ tô màu đen. Nếu có nhiều đáp án, hãy in ra đáp án bất kỳ.

*   Nếu không thể tô màu sao cho đảm bảo $M$ điều kiện, hãy in ra $-1$
.

### **Scoring**

*   Subtask 1 (30% số điểm): $N,M≤20$
*   Subtask 2 (20% số điểm): $c∈{0,2}$ và $N,M≤1000$
*   Subtask 3 (50% số điểm): Không còn ràng buộc nào khác

### **Example**

**input**

4 4

1 2 2

2 3 1

3 4 1

4 1 2

**output**

1 1 0 1


**input**

5 5

1 2 1

2 3 1

2 4 1

2 5 1

4 5 1


**output**

-1

**input**

4 5

1 2 1

2 3 0

2 4 1

3 1 1

3 4 1


**output**

1 0 0 1

**input**

2 1

1 2 1

**output**

1 0

### **Note**

Với ví dụ $4$, cách tô màu [$0$ $1$] cũng thoả mãn.
