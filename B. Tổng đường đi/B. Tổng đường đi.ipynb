# B. Tổng đường đi

time limit per test: 2 seconds

memory limit per test: 128 megabytes

Như bạn đã biết, cây là một đồ thị liên thông gồm $N$ đỉnh và $N-1$ cạnh. Cây cũng có một tính chất là có chính xác một đường đi đơn giản giữa hai đỉnh bất kỳ. Một đường đi đơn giản là đường đi không có đỉnh nào được lặp lại.

Bạn được cung cấp một cây, với mỗi cạnh được gắn với một trọng số không âm. Trọng số của một đường đi là tích của tất cả trọng số của các cạnh trên đường đi đó. Trọng số của cây là tổng trọng số của tất cả các đường đi. Những đường đi đi theo hướng ngược nhau (ví dụ $A$ tới $B$ và $B$ tới $A$) được coi là giống nhau, nên khi tính toán trọng số của cây, chúng chỉ được đếm một lần duy nhất.

Bạn hãy viết một chương trình, tính trọng số của cây. Đáp án có thể rất lớn, hãy in ra sau khi chia dư cho $1000000007$.

### **Input**

*   Dòng đầu tiên gồm số nguyên dương $N$ $(2≤N≤100000)$
*   $N-1$ dòng tiếp theo, mỗi dòng gồm ba số nguyên dương $A,B,W$ $(1≤A,B≤N, 0≤W≤1000)$

### **Output**

*   In ra trọng số của cây sau khi chia dư cho $1000000007$.

### **Scoring**

*   Subtask 1 (30% số điểm): $N≤1000$ và mỗi đỉnh có tối đa 2 đỉnh kề.
*   Subtask 2 (30% số điểm): $N≤100000$ và mỗi đỉnh có tối đa 2 đỉnh kề.
*   Subtask 3 (40% số điểm): Không còn ràng buộc nào khác

### **Examples**

**input**

3

3 2 100

2 1 100

**output**

10200

**input**

4

1 2 5

1 3 5

1 4 5

**output**

90

**input**

5

1 2 2

2 3 3

4 3 2

5 3 2

**output**

55

### **Note**


Giải thích ví dụ 1:

*   Đường $1→2$ có trọng số là $100$.

*   Đường $2→3$ có trọng số là $100$.

*   Đường $1→3$ có trọng số là $10000$.
